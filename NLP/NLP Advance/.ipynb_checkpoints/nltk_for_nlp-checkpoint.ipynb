## What is nltk

Natural Language Toolkit

NLTK is a leading for Resource and Developing Nautral Programming Language on Python programing language. It provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordNet, along with a suite of text processing libraries for classification, tokenization, stemming, tagging, parsing, and semantic reasoning, wrappers for industrial-strength NLP libraries, and an active discussion forum.

### stopwords 

All Natural Languages have some words for making sentence good. Example a, an, the, end.... But in data analyasis for machine those are words are not any meanning. 

```python
from nltk.corpus import stopwords
```

```python
# check stopwords
stop_words = stopwords.words('English')
```

```python
words = word_tokenize(text)
```

```python
# extract stopwords from words
filtred_words = [word for word in words if word not in stop_words]
```
